```python
from utils import load_json_from_url, load_config

cfg = load_config("config.yaml")
counterfact_data = load_json_from_url(cfg['path']['counterfact_path'])
```

```python
cf_dialogue = []
rel = set()
for data in counterfact_data:
    dt = {}
    sro = data['requested_rewrite']
    dt['case_id'] = data['case_id']
    dt['relation_id'] = sro['relation_id']
    dt['subject'] = sro['subject']
    dt['target_true'] = sro['target_true']['str']
    dt['target_new'] = sro['target_new']['str']
    dt['cloze'] = sro['prompt'].format(sro['subject'])
    dt['cloze_prompt'] = sro['prompt']
    rel.add(sro['relation_id'])
    cf_dialogue.append(dt)
    
    
```

```python
import json

def save_list_to_jsonl(data_list, filepath):
    with open(filepath, 'w') as file:
        for item in data_list:
            file.write(json.dumps(item) + '\n')

# 사용 예시

filepath = 'counterfact_memit.jsonl'
save_list_to_jsonl(counterfact_data, filepath)
```
