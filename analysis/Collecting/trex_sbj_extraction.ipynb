```python
import pandas as pd
df_trx = pd.read_excel('df_imsi_trex.xlsx')
df_sbj = pd.read_excel('df_one_hop_sbj.xlsx')
df_sbj_last = pd.read_excel('df_one_hop_sbj_last.xlsx')
print(len(df_trx))
print(len(df_sbj))
print(len(df_sbj_last))
```

```python
df_sbj.head(20)
```

```python
from tqdm import tqdm
for i in tqdm(range(len(df_trx))):
    idx = df_trx.loc[i, 'index']
    for j in range(len(df_sbj)):
        if df_sbj.loc[j, 'index'] == idx:
            words = df_trx.loc[i, 'trex_one_hop']
            if not pd.isna(words):
                df_sbj.loc[j, 'subject_id'] = df_trx.loc[i, 'TREx_id']
                df_sbj.loc[j, 'one_hop'] = words
            break
```
