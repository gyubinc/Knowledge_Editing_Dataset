```python
import pandas as pd

df = pd.read_excel("exact_qid_match.xlsx")

df.head()
```

```python
good = list()
bad = list()

for i in range(len(df)):
    if df['subject'][i] == df['wiki_subject'][i]:
        good.append(i)
    else:
        bad.append(i)
        
print(len(good))
print(len(bad))
```

```python
df_good = df.loc[good].reset_index(inplace = False, drop = True)
df_bad = df.loc[bad].reset_index(inplace = False, drop = True)
```
