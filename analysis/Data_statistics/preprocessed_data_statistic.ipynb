```python
import pandas as pd

df = pd.read_excel("../../data/one_hop/preprocessed_df.xlsx")
df.fillna('null', inplace = True)
df
```

```python
count = 0
for i in range(len(df)):
    if (len(df.loc[i, 'sbj_one_hop'].split(',')) != df.loc[i, 'sbj_hop_num'] or len(df.loc[i, 'obj_one_hop'].split(',')) != df.loc[i, 'obj_true_hop_num'] or len(df.loc[i, 'obj_new_one_hop'].split(',')) != df.loc[i, 'obj_new_hop_num']):
        count += 1
print(count)
```

```python
check = ["sbj_hop_num", 'obj_true_hop_num', 'obj_new_hop_num']

def stat(df, word, k):
    count = 0
    word_list = list(df[word])
    for freq in word_list:
        if freq >= k:
            count += 1
    print(f'Word count : {count} ({round(count / len(word_list)*100,2)}%) with {word} >= {k}')
    return count

for i in range(0, 11):
    stat(df, check[0], i)
    
for i in range(0, 11):
    stat(df, check[1], i)

for i in range(0, 11):
    stat(df, check[2], i)
```

```python
check = ["sbj_hop_num", 'obj_true_hop_num', 'obj_new_hop_num']

def stat2(df, word, k):
    count = 0
    sbj_list = list(df[word[0]])
    obj_list = list(df[word[1]])
    obj_new_list = list(df[word[2]])
    for i in range(len(sbj_list)):
        if sbj_list[i] >= k and obj_list[i] >= k and obj_new_list[i] >= k:
            count += 1
    print(f'Word count : {count} ({round(count / len(sbj_list)*100,2)}%) with triplet >= {k}')
    return count

for i in range(0, 11):
    stat2(df, check, i)
```
