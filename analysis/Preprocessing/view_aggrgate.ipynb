```python
import pandas as pd

sbj_hop = pd.read_excel('../../data/one_hop/df_sbj_full_prep_241110.xlsx')
obj_true_hop = pd.read_excel('../../data/one_hop/df_obj_true_full_prep_241110.xlsx')
obj_new_hop = pd.read_excel('../../data/one_hop/df_obj_new_full_prep_241110.xlsx')
views = pd.read_excel('../../data/view.xlsx')
views.head()
```

```python
sbj_hop.head()
```

```python
obj_true_hop.head()
```

```python
df_total = pd.concat([sbj_hop, obj_true_hop, obj_new_hop, views], axis = 1)
columns_to_keep = [i for i in range(len(df_total.columns)) if i not in [5, 10, 15, 16]]
df_total = df_total.iloc[:, columns_to_keep]
df_total.head()
```

```python
df_total.columns.values[3] = 'sbj_one_hop'
df_total.columns.values[4] = 'sbj_hop_num'
df_total.columns.values[5] = 'obj_true'
df_total.columns.values[6] = 'obj_true_id'
df_total.columns.values[8] = 'obj_true_hop_num'
df_total.columns.values[9] = 'obj_new'
df_total.columns.values[10] = 'obj_new_id'
df_total.columns.values[12] = 'obj_new_hop_num'
df_total.columns.values[13] = 'sbj_views'

df_total.head()
```

```python
df_total.to_excel('../../data/preprocessed_df_241110.xlsx', index = False)
```
