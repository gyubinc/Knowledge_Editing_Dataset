```python
import json

with open('../../data/TREx/answer.json', 'r', encoding = 'utf-8') as file:
    data = json.load(file)
          
print(data)
```

```python
from collections import Counter

# value 기준으로 개수 세기
value_counts = Counter(data.values())

# 결과 출력
# 내림차순 정렬된 결과 출력
# for value, count in value_counts.most_common():
#     print(f"Value {value} appears {count} times")
    
    # 등장 횟수가 2회 이상인 value들을 리스트로 만들기
duplicate_values = [value for value, count in value_counts.items() if count > 1]

print("Duplicate values:", duplicate_values)
```

```python
from collections import defaultdict


# value를 key로, 해당 value를 가진 key들을 리스트로 모으기
reversed_dict = defaultdict(list)
for key, value in data.items():
    reversed_dict[value].append(key)

# 리스트 크기가 2 이상인 경우만 필터링하여 출력
filtered_dict = {value: keys for value, keys in reversed_dict.items() if len(keys) >= 2}

print(filtered_dict)
```

```python
import requests

def get_qid_exact_label(text):
    url = 'https://query.wikidata.org/sparql'
    query = '''
    SELECT ?item WHERE {
      ?item rdfs:label "%s"@ko .
    }
    ''' % text
    headers = {
        'Accept': 'application/sparql-results+json'
    }
    params = {
        'query': query,
        'format': 'json'
    }
    response = requests.get(url, headers=headers, params=params)
    if response.status_code != 200:
        raise Exception("SPARQL 쿼리에 실패했습니다: 상태 코드 %d" % response.status_code)
    data = response.json()
    results = data['results']['bindings']
    if len(results) == 1:
        qid_url = results[0]['item']['value']
        qid = qid_url.split('/')[-1]
        return qid
    else:
        return None

# 사용 예시
text = "Seoul"
qid = get_qid_exact_label(text)
print(qid == None)  # 출력 예시: Q937
```
